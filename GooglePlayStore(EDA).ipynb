```python
def size_process(item):
    if str(item)[-1] == "M":
        result = float(str(item).replace("M",""))
        result = result * 1024
        return result
    elif str(item)[-1] == "K":
        result = float(str(item).replace("K",""))
        return result
    else:
        return str(np.NAN)

size_process('19M')
```

```python
def clean_version(ver):
    # Handle NaN or empty string
    if pd.isna(ver) or str(ver).strip() == "":
        return ver
    
    ver = str(ver).strip()
    parts = ver.split(" - ")
    cleaned_parts = []

    for part in parts:
        # Extract only the major.minor version
        match = pd.Series(part.strip()).str.extract(r"(\d+\.\d+)")
        if not match.empty and not pd.isna(match[0][0]):
            cleaned_parts.append(match[0][0])

    if len(cleaned_parts) == 0:
        return np.nan  # fallback if nothing matched
    elif len(cleaned_parts) == 1:
        return cleaned_parts[0]  # only one version, return it
    else:
        # Compare and return the higher version
        # Convert version strings to tuples for comparison
        versions_as_tuples = [tuple(map(int, v.split('.'))) for v in cleaned_parts]
        max_version = max(versions_as_tuples)
        return ".".join(map(str, max_version))
```

```python
df_copy["Android Ver"] = df_copy["Android Ver"].apply(clean_version)
```
